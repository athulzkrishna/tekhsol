```python
import openpyxl
import glob, os
os.chdir("C:\\Users\\Athul krishna\\myexcels\\production\\")
for file in glob.glob("*.xlsx"):
    print(file)
#q[file[4:6]]=(extracty(file,21,l))    
#print(q)
q=[]
q=extracty(file,12,q)
print(q)
```

```python
#Finding the mean and SD
import statistics
mean_of_production=statistics.mean(q)
mean_of_stop=statistics.mean(v)
dev_production=statistics.stdev(q)
dev_stop=statistics.stdev(v)
print(mean_of_production)
print(mean_of_stop)
print(dev_production)
print(dev_stop)
print(min(v))
print(max(v))
```

```python
import random

prod=[]
stopp=[]
#prod=dataset11+dataset1
#stopp=dataset22+dataset2
#random.shuffle(prod)
#random.shuffle(stopp)

for i in range(100):
    if q[i]>mean_of_production:
        prod.append(mean_of_production+(random.randint(0,int(dev_production))))
    else:
        prod.append(mean_of_production-(random.randint(0,int(dev_production))))
for i in range(100):
    if v[i]>mean_of_stop:
        stopp.append(mean_of_stop+(random.randint(0,int(dev_stop))))
    else:
        stopp.append(mean_of_stop-(random.randint(0,int(dev_stop))))
print(prod)
print(stopp)       
```
